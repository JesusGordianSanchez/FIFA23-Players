```python
#define table
df = tabs[0]

#list all column names of table
list(df)
```

```python
#filter Dataframe to only contain five columns
df_final = df.iloc[:,[3,0,10,13,11]]

#rename columns
df_final.columns = ["Player", "Overall", "Skill Moves","Foot","Awareness",]

#view first few rows of final Dataframe
print(df_final.head(50))
```

```python
#Separate the "player" and "position' from the same column.

# df_final[["Player Name", "Position"]]= df_final["Player"].str.rsplit(' ', n=1,expand=True)

df_final.loc[:, 'Player Name'] = df_final['Player'].str.rsplit(' ', n = 1).str[0]
df_final.loc[:, 'Position'] = df_final['Player'].str.rsplit(' ', n = 1).str[1]
```

```python
df_final.drop(columns=['Player'], inplace=True)
```

```python
df_final
```

```python
#Add another column to contain secuencial numbers from 1 to 50

n = 50
df_final['No.'] = range(1, n+1)
#Ddf_final
```

```python
#Delete columns ("Nueva Columna")

df_final = df_final.drop(columns=['Nueva_Columna'])
df_final
```

```python
# Get the current order of the columns
column_order = df_final.columns.tolist()

# Change the order of columns
new_column_order = ["No.", "Player Name", "Foot", "Awareness", "Position", "Skill Moves", "Overall"]

# Reassign the columns in the new order
df_final = df_final[new_column_order]
df_final
```
